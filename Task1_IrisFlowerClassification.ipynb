## **Importing Necessary Libraries.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import LabelEncoder
```

## **Loading the dataset.**

You can read the dataset in form of a csv file or directly in colab using the command **" from sklearn.datasets import load_iris "** both are same. Here I will import my csv file. 

```python
dataset = pd.read_csv("iris.csv")
```

```python
dataset.shape
```

```python
dataset.head
```

### **Adding coloumn labels.**

```python
dataset.columns = ['sepal length','sepal width','petal length','petal width','class']
```

```python
dataset.head()
```

## **Describing the dataset.**

```python
print(dataset.describe())
```

## **Checking for null values if any.**

```python
print(dataset.isnull().sum())
```

## **Visualising the data by ploting.** 

```python
sns.boxplot(data=dataset, width=0.5, fliersize=5)
sns.set(rc = {"figure.figsize" : (6,6)})
```

```python
Features = ['sepal length','sepal width','petal length','petal width']
X = dataset[Features].values
```

**We will here Label Encode coloumn "class".**
Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form.

```python
le = LabelEncoder()
dataset['class'] = le.fit_transform(dataset['class'])
y = dataset['class'].values
```

```python
y
```

```python
X
```

```python
#visualise the data by ploting
plt.plot(X[:,0][y == 0]*X[:,1][y == 0],X[:,2][y == 0]*X[:,3][y == 0],'r.', label = 'Setosa')
plt.plot(X[:,0][y == 1]*X[:,1][y == 1],X[:,2][y == 1]*X[:,3][y == 1],'g.', label = 'Versicolour')
plt.plot(X[:,0][y == 2]*X[:,1][y == 2],X[:,2][y == 2]*X[:,3][y == 2],'b.', label = 'Virginica')

plt.legend()
plt.show()
```

Here we can clearly see that **Iris-Sectosa** is clearly seperable by the other two based on areas of petal and sepal.

## **Standardize the data.**

```python
#standardize the data
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

## **Divide the data for Training and Testing.**

```python
#Divide data into train and test 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

## **Using the Machine Learning Algorithm.**
**Here we are using the Logistic Regression.**

```python
#Using ML algorithm to predict the calssification
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

## **Training the ML model.**

```python
log_reg.fit(X_train, y_train)
```

## **Testing the ML model.**
Using test data.

```python
print("Testing Accuracy : ",log_reg.score(X_test,y_test)*100)
```

# **Testing the ML model on whole Dataset.**

```python
print("Accuracy : ",log_reg.score(X,y)*100)
```
